```python
import numpy as np
```

```python
def make_patients(n=400):
    # low-risk: tight cluster near the centre
    inner = np.random.randn(n//2, 2) * 0.5
    # high-risk: a ring around the cluster
    angles = np.random.rand(n//2) * 2*np.pi
    radius = 2.0 + np.random.randn(n//2) * 0.3
    outer  = np.c_[radius*np.cos(angles),
                   radius*np.sin(angles)]
    X = np.vstack([inner, outer])      # (400, 2)
    y = np.array([0]*200 + [1]*200).reshape(-1, 1)
    return X, y
```

```python
X, y = make_patients(n = 400)
```
